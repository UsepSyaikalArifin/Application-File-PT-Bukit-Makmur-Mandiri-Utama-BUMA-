Status Code
Ada beberapa jenis status code yang umumnya dijumpai, antara lain:

200. Artinya permintaan sukses dipenuhi.
404. Artinya berkas yang diminta tidak dapat ditemukan.
500. Artinya akses permintaan ditolak.
501. Artinya terjadi kesalahan pada server.

## Apa Kabar Jawa Barat?

```python
import requests as req
# import pandas as pd
resp_jabar = req.get("https://data.covid19.go.id/public/api/prov_detail_JAWA_BARAT.json")
# resp_jabar_1 = pd.read_json("https://data.covid19.go.id/public/api/prov_detail_JAWA_BARAT.json")
cov_jabar_raw = resp_jabar.json()
# resp_jabar_1 = pd.DataFrame(cov_jabar_raw)
display(cov_jabar_raw)
# display(resp_jabar_1)
```

## Memahami Kasus COVID-19 di Jawa Barat

```python
print("Nama - nama elemen Utama : ")
display(cov_jabar_raw.keys())
print("Jumlah total kasus COVID-19 di Jawa Barat : ")
display(cov_jabar_raw["kasus_total"])
print("Persentase kematian akibat COVID-19 di Jawa Barat : ")
display(cov_jabar_raw["meninggal_persen"])
print("Persentase tingkat kesembuhan dari COVID-19 di Jawa Barat : ")
display(cov_jabar_raw["sembuh_persen"])
```

## Memperoleh Informasi yang Lebih Lengkap

```python
import numpy as np
import pandas as pd

cov_jabar = pd.DataFrame(cov_jabar_raw["list_perkembangan"])
print("Info cov_jabar : ")
display(cov_jabar.info())
print("5 Data teratas cov_jabar : ")
display(cov_jabar.head())
```

## Menjinakkan Data

```python
cov_jabar_tidy = (cov_jabar.drop(
                                 columns=[
                                            item for item in cov_jabar.columns
                                            if item.startswith("AKUMULASI") or item.startswith("DIRAWAT")
                                          ]
                                )
                                .rename(columns=str.lower)
                                .rename(columns={"kasus" : "kasus_baru"})
                  )

cov_jabar_tidy["tanggal"] = pd.to_datetime(cov_jabar_tidy["tanggal"]*1e6, unit="ns")
display(cov_jabar_tidy.head())
```

## Visualisasi Data - 1

```python
import matplotlib.pyplot as plt

plt.clf()
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
ax[0].plot(cov_jabar_tidy["tanggal"], cov_jabar_tidy["kasus_baru"])
ax[1].bar(data=cov_jabar_tidy, x="tanggal", height="kasus_baru", color="red")
plt.show()
```

## Visualisasi Data - 2

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.clf()
fig, ax = plt.subplots(3, 1, figsize=(10, 15))
ax[0].bar(data=cov_jabar_tidy, x="tanggal", height="kasus_baru", color="red")
ax[1].bar(data=cov_jabar_tidy, x="tanggal", height="sembuh", color="blue")
ax[2].bar(data=cov_jabar_tidy, x="tanggal", height="meninggal", color="green")
fig.suptitle("Kasus Harian Positif COVID-19 di Jawa Barat", y=1, fontsize=20, fontweight="bold", ha="center")
ax[0].set_title("Terjadi pelonjakan kasus di awal bulan Juli akibat klaster Secapa AD Bandung", fontsize=12)
ax[1].set_title("Kasus Harian Sembuh Dari COVID-19 di Jawa Barat", fontsize=12)
ax[2].set_title("Kasus Harian Meninggal Dari COVID-19 di Jawa Barat", fontsize=12)
ax[0].set_xlabel("Waktu", fontsize=12, fontweight="bold")
ax[1].set_xlabel("Waktu", fontsize=12, fontweight="bold")
ax[2].set_xlabel("Waktu", fontsize=12, fontweight="bold")
ax[0].set_ylabel("Jumlah Kasus", fontsize=12, fontweight="bold")
ax[1].set_ylabel("Jumlah Kasus Sembuh", fontsize=12, fontweight="bold")
ax[2].set_ylabel("Jumlah Kasus Meninggal", fontsize=12, fontweight="bold")
# ax[0].text(1, -0.19, "Sumber data : covid.19.go.id", color="blue", ha="right", transform=ax[0].transAxes)
# ax[1].text(1, -0.19, "Sumber data : covid.19.go.id", color="blue", ha="right", transform=ax[0].transAxes)
# ax[2].text(1, -0.19, "Sumber data : covid.19.go.id", color="blue", ha="right", transform=ax[0].transAxes)
# ticks_loc = ax[0].get_xticks().tolist()
# ax[0].xaxis.set_major_locator(mdates.MonthLocator())
# ax[0].set_xticklabels(["{}".format(x) for x in ticks_loc])

ax[0].set_xticklabels(ax[0].get_xticks(), rotation=90)
ax[0].xaxis.set_major_locator(mdates.MonthLocator())
ax[0].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

ax[1].set_xticklabels(ax[1].get_xticks(), rotation=90)
ax[1].xaxis.set_major_locator(mdates.MonthLocator())
ax[1].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

ax[2].set_xticklabels(ax[2].get_xticks(), rotation=90)
ax[2].xaxis.set_major_locator(mdates.MonthLocator())
ax[2].xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
