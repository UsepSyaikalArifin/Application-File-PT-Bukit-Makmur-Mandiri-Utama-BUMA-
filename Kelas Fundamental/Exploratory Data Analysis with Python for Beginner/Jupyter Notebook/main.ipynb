```python
order_df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/order.csv")

numeric_columns = order_df.columns[order_df.dtypes != "object"].to_list()

for i, column in enumerate(numeric_columns):
    print("Mean colomn {} = {}".format(column, order_df.loc[:,column].mean()))
    print("Median colomn {} = {}".format(column, order_df.loc[:,column].median()))
    print("Modus colomn {} = {}".format(column, order_df.loc[:,column].mode()))
    print("Std colomn {} = {}".format(column, order_df.loc[:,column].std()))
    print("Var colomn {} = {}\n".format(column, order_df.loc[:,column].var()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/order.csv")

numeric_columns = order_df.columns[order_df.dtypes != "object"].to_list()

fig, axs = plt.subplots(4, 1, figsize=(12, 8))
for i, column in enumerate(numeric_columns):
    color = sns.color_palette("binary_r")
    sns.histplot(data=order_df, x=column, kde=True, ax=axs[i], palette=color)
plt.tight_layout()
plt.show()
```

## Groupby -> Summary

```python
import pandas as pd

order_df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/order.csv")

display(order_df.head())

rata_rata_1 = order_df["price"].groupby(order_df["payment_type"]).mean().round()
rata_rata_2 = order_df.groupby(["payment_type"])["price"].mean().round()

display(rata_rata_1)
display(rata_rata_2)
```

## Tugas Praktek

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

order_df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/order.csv")

median_price = order_df.groupby(["payment_type"])["price"].median()
df_median_price = pd.DataFrame(median_price)

order_df.rename(columns={"freight_value":"shipping_cost"}, inplace=True)

sort_value = order_df.sort_values(by="shipping_cost", ascending=False)

std_value = order_df.groupby(["product_category_name"])["product_weight_gram"].std()
df_std_value = pd.DataFrame(std_value)

display(df_median_price.sort_values(by="price", ascending=True))
display(sort_value.head())
display(df_std_value)

fig, axs = plt.subplots(1, 1, figsize=(12, 8))
sns.histplot(data=order_df, x="quantity", bins=order_df["quantity"].nunique())
plt.tight_layout()
plt.show()
```
