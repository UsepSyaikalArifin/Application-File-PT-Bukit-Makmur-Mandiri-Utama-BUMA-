```python
import datetime

dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv")

dataset["order_month"] = dataset["order_date"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime("%Y-%m"))

print("Informasi dataset : ")
display(dataset.info())
display(dataset.head())
```

```python
dataset["gmv"] = dataset["quantity"] * dataset["item_price"]

display(dataset.info())
display(dataset.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

monthly_amount = dataset.groupby(["order_month"])["gmv"].sum().reset_index()

fig, axs = plt.subplots(1, 1, figsize=(12, 5))
sns.lineplot(data=monthly_amount, x="order_month", y="gmv",color="green", marker="D", linestyle="-.", linewidth=3)
axs.set_title("Grafik order_month vs gmv", fontsize=20, fontweight="bold",color="blue", loc="center", pad=10)
axs.set_xlabel("order_month", fontsize=15)
axs.set_ylabel("gmv (Miliar)", fontsize=15)
axs.set_ylim(0)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
axs.text(3, 7200000000, "The GMV Increased Significantly on October 2019-10", fontsize=13, ha='center', va='center', color="red")
plt.grid(color="darkgray", linestyle=":", linewidth=0.5)
plt.tight_layout()
# plt.savefig("../File Image/Gambar 3.png", facecolor="white", dpi = 300)
plt.show()
```

```python
import numpy as np

daily_number_customers_on_dec = dataset[dataset["order_month"] == "2019-12"].groupby(["order_date"])["customer_id"].nunique().reset_index()

fix, axs = plt.subplots(1, 1, figsize=(10, 5))
sns.lineplot(data=daily_number_customers_on_dec, x="order_date", y="customer_id", marker="D",markersize=6, linewidth=2)
axs.set_title("Daily Number of Customers - December 2019", fontsize=20, fontweight="bold", color="orange", pad=20)
axs.set_xlim(-1, 27)
axs.set_ylim(10, 50)
stars, ends = axs.get_xlim()
axs.set_xticks(np.arange(stars, ends, 4))
axs.set_xlabel("Order Date", fontsize=15)
axs.set_ylabel("Customer ID", fontsize=15)
plt.grid(color="darkgray")
plt.tight_layout()
plt.savefig("../File Image/Image", facecolor="white", dpi=300)
plt.show()
```
