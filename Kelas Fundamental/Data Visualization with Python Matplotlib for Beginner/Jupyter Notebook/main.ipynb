```python
import datetime

dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv")

dataset["order_month"] = dataset["order_date"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime("%Y-%m"))

print("Informasi dataset : ")
display(dataset.info())
display(dataset.head())
```

```python
dataset["gmv"] = dataset["quantity"] * dataset["item_price"]

display(dataset.info())
display(dataset.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

monthly_amount = dataset.groupby(["order_month"])["gmv"].sum().reset_index()

fig, axs = plt.subplots(1, 1, figsize=(12, 5))
sns.lineplot(data=monthly_amount, x="order_month", y="gmv",color="green", marker="D", linestyle="-.", linewidth=3)
axs.set_title("Grafik order_month vs gmv", fontsize=20, fontweight="bold",color="blue", loc="center", pad=10)
axs.set_xlabel("order_month", fontsize=15)
axs.set_ylabel("gmv (Miliar)", fontsize=15)
axs.set_ylim(0)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
axs.text(3, 7200000000, "The GMV Increased Significantly on October 2019-10", fontsize=13, ha='center', va='center', color="red")
plt.grid(color="darkgray", linestyle=":", linewidth=0.5)
plt.tight_layout()
# plt.savefig("../File Image/Gambar 3.png", facecolor="white", dpi = 300)
plt.show()
```

```python
import numpy as np

daily_number_customers_on_dec = dataset[dataset["order_month"] == "2019-12"].groupby(["order_date"])["customer_id"].nunique().reset_index()

fix, axs = plt.subplots(1, 1, figsize=(10, 5))
sns.lineplot(data=daily_number_customers_on_dec, x="order_date", y="customer_id", marker="D",markersize=6, linewidth=2)
axs.set_title("Daily Number of Customers - December 2019", fontsize=20, fontweight="bold", color="orange", pad=20)
axs.set_xlim(-1, 27)
axs.set_ylim(10, 50)
stars, ends = axs.get_xlim()
axs.set_xticks(np.arange(stars, ends, 4))
axs.set_xlabel("Order Date", fontsize=15)
axs.set_ylabel("Customer ID", fontsize=15)
plt.grid(color="darkgray")
plt.tight_layout()
plt.savefig("../File Image/Image", facecolor="white", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sum_of_gmv_baseon_brands = dataset.groupby(["order_month","brand"])["gmv"].sum().unstack()
fig, axs = plt.subplots(1, 1, figsize=(13,5))
sns.lineplot(data=sum_of_gmv_baseon_brands, marker="D", linestyle="-", linewidth=2, ax=axs, palette="Set1")
axs.legend(loc="best", bbox_to_anchor=(1, 1), title="Brands", title_fontsize=15,ncol=2)
axs.set_title("Order Month vs Brand", fontsize=20, fontweight="bold")
axs.set_xlabel("Order Month", fontsize=15)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
axs.set_ylabel("GMV (Miliar)", fontsize=15)
plt.grid(color="darkgray")
plt.tight_layout()
plt.show()
```

```python
top_provinces = dataset.groupby(["province"])["gmv"].sum().reset_index().sort_values(by="gmv", ascending=False).head(5)
display(top_provinces)
```

```python
dataset["province_top"] = dataset["province"].apply(lambda x : x if x in top_provinces["province"].to_list() else "other")

display(dataset.head())
```

```python
sum_of_gmv_baseon_province_top = dataset.groupby(["order_month","province_top"])["gmv"].sum().unstack()

fig, axs = plt.subplots(1, 1, figsize=(13,5))
sns.lineplot(data=sum_of_gmv_baseon_province_top , marker="D", linestyle="-", linewidth=2, ax=axs, palette="plasma")
axs.legend(loc="best", bbox_to_anchor=(1, 1), title="Brands", title_fontsize=15,ncol=2)
axs.annotate(
             "DKI Jakarta Mendominasi",
             xy=(3, 3455000000),
             xytext=(0.7, 3700000000),
             weight="bold",
             color="red",
             ha="center",
             va="center",
             arrowprops=dict(
                              arrowstyle="->",
                              connectionstyle="angle",
                              color="red"
                            )
            )
axs.annotate(
             "GMV Other Meningkat Pesat",
             xy=(5, 1150000000),
             xytext=(3, 2000000000),
             weight="bold",
             color="red",
             ha="center",
             va="center",
             arrowprops=dict(
                              arrowstyle="->",
                              connectionstyle="arc3",
                              color="red"
                            )
            )
axs.set_title("Order Month vs Brand", fontsize=20, fontweight="bold")
axs.set_xlabel("Order Month", fontsize=15)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
axs.set_ylabel("GMV (Miliar)", fontsize=15)
plt.grid(color="darkgray")
plt.tight_layout()
plt.show()
```

```python
dataset_dki_q4 = dataset[(dataset["province"] == "DKI Jakarta") & (dataset["order_month"] >= "2019-10")]
display(dataset_dki_q4.head())
```

```python
gmv_per_city_dki_q4 = dataset_dki_q4.groupby(["city"])["gmv"].sum().reset_index()

explode = (0, 0, 0.1, 0, 0)
fig, axs = plt.subplots(1, 1, figsize=(9, 9))
plt.pie(data=gmv_per_city_dki_q4, x="gmv", labels="city", autopct="%0.1f%%", pctdistance=0.5, explode=explode, shadow=True,textprops=dict(color="w",
                                                                                                             fontweight="bold",
                                                                                                             fontsize=15))
axs.set_title("Pie Chart", fontsize=20, fontweight="bold", color="white")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

display(gmv_per_city_dki_q4.head())

fig, axs = plt.subplots(1, 1, figsize=(10, 6))
sns.barplot(data=gmv_per_city_dki_q4.sort_values(by="gmv",ascending=False), x="city", y="gmv")
axs.set_title("Komposisi GMV Top City", fontsize=25, fontweight="bold")
axs.set_xlabel("City", fontsize=15)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
for p in axs.patches:
    axs.annotate(
                 "{}".format(p.get_height()),
                 xy=(p.get_x() + p.get_width()/2 , p.get_height()),
                 fontsize=12,
                 ha="center",
                 va="center",
                 xytext=(0, 10),
                 textcoords="offset points"
                )
axs.axes.yaxis.set_visible(False)
sns.despine(left=True, top=True, right=True)
plt.tight_layout()
plt.show()
```

```python
gmv_city_baseon_order_month = dataset_dki_q4.groupby(["city", "order_month"])["gmv"].sum().reset_index()

import matplotlib.pyplot as plt
import seaborn as sns

display(gmv_city_baseon_order_month)

fig, axs = plt.subplots(1, 1, figsize=(10, 6))
sns.barplot(data=gmv_city_baseon_order_month.sort_values(by="gmv", ascending=False), x="city", y="gmv", hue="order_month")
axs.set_title("Komposisi GMV Top City", fontsize=25, fontweight="bold")
axs.set_xlabel("City", fontsize=15)
axs.set_yticks(axs.get_yticks())
axs.set_yticklabels(axs.get_yticks()/1000000000)
for p in axs.patches:
    axs.annotate(
                 "{0:.2f}".format((p.get_height()/100000000).astype(float)),
                 xy=(p.get_x() + p.get_width()/2 , p.get_height()),
                 fontsize=12,
                 ha="center",
                 va="center",
                 xytext=(0, 10),
                 textcoords="offset points"
                )
axs.axes.yaxis.set_visible(False)
sns.despine(left=True, top=True, right=True)
plt.tight_layout()
plt.show()
```
