## Eksplorasi Data: Memahami Data dengan Visual

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_revenue_weekend = ["Revenue","Weekend"]
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
for i, column in enumerate(columns_revenue_weekend):
    colors = sns.color_palette("Set{}".format(i+1))
    sns.countplot(x=column, data=dataset, palette=colors, ax=axs[i])
    axs[i].set_title(column, fontsize=20, fontweight="bold")
    axs[i].set_xlabel(column, fontsize=15)

    for p in axs[i].patches:
        axs[i].annotate(
                        format(p.get_height()),
                        xy = (p.get_x() + p.get_width()/2, p.get_height()),
                        ha = "center",
                        va = "center",
                        xytext = (0, 10),
                        textcoords = "offset points"
                       )
    sns.despine(right=True, top=True, left=True)
    axs[i].axes.yaxis.set_visible(False)
plt.tight_layout()
plt.show()
```

## Tugas Praktek

```python
def count_region(dataset, x):
    fig, axs = plt.subplots(1, 1, figsize=(12, 5))
    sns.countplot(data=dataset, x=x, palette=sns.color_palette("tab10"))
    axs.set_title("Distribution of Customers", fontsize=20, fontweight="bold")
    axs.set_xlabel(x, fontsize=15)
    for p in axs.patches:
        axs.annotate(
                        "{} User".format(p.get_height()),
                        xy = (p.get_x() + p.get_width()/2, p.get_height()),
                        ha = "center",
                        va = "center",
                        xytext = (0, 10),
                        textcoords = "offset points"
                       )
    sns.despine(right=True, top=True, left=True)
    axs.axes.yaxis.set_visible(False)
    plt.tight_layout()
    plt.show()

count_region(dataset, "Region")
```

## Handling Missing Value -> Drop All Rows Missing Values

```python
def dataframe_missing_values(dataset):
    df_columns = dataset.isnull().sum().index
    df_values = dataset.isnull().sum().values
    df_missing_value = pd.DataFrame({
                                        "Columns" : df_columns,
                                        "Missing_Values" : df_values
                                    })
    return df_missing_value

def dropall_missing_value(dataset):
    dataset_clean = dataset.dropna()
    df_clean_missing_value = dataframe_missing_values(dataset_clean)
    return df_clean_missing_value

missing_values = dataframe_missing_values(dataset)
clean_dataset = dropall_missing_value(dataset)

display(missing_values)
display(clean_dataset)
```

## Handling Missing Value -> Impute Data Point Missing Values

```python
def impute_missing_value_with_mean(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].mean(), inplace=True)
    return dataset
    
def impute_missing_value_with_median(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].median(), inplace=True)
    return dataset

dataset_clean_with_mean = impute_missing_value_with_mean(dataset)
dataset_clean_with_median = impute_missing_value_with_mean(dataset)
dataset_clean_with_mean = dataframe_missing_values(dataset_clean_with_mean)
dataset_clean_with_median= dataframe_missing_values(dataset_clean_with_median)

display(dataset_clean_with_mean)
display(dataset_clean_with_median)
```
