## Tugas Praktek

```python
def count_region(dataset, x):
    fig, axs = plt.subplots(1, 1, figsize=(12, 5))
    sns.countplot(data=dataset, x=x, palette=sns.color_palette("tab10"))
    axs.set_title("Distribution of Customers", fontsize=20, fontweight="bold")
    axs.set_xlabel(x, fontsize=15)
    for p in axs.patches:
        axs.annotate(
                        "{} User".format(p.get_height()),
                        xy = (p.get_x() + p.get_width()/2, p.get_height()),
                        ha = "center",
                        va = "center",
                        xytext = (0, 10),
                        textcoords = "offset points"
                       )
    sns.despine(right=True, top=True, left=True)
    axs.axes.yaxis.set_visible(False)
    plt.tight_layout()
    plt.show()

count_region(dataset, "Region")
```

## Handling Missing Value -> Drop All Rows Missing Values

```python
def dataframe_missing_values(dataset):
    df_columns = dataset.isnull().sum().index
    df_values = dataset.isnull().sum().values
    df_missing_value = pd.DataFrame({
                                        "Columns" : df_columns,
                                        "Missing_Values" : df_values
                                    })
    return df_missing_value

def dropall_missing_value(dataset):
    dataset_clean = dataset.dropna()
    df_clean_missing_value = dataframe_missing_values(dataset_clean)
    return df_clean_missing_value

missing_values = dataframe_missing_values(dataset)
clean_dataset = dropall_missing_value(dataset)

display(missing_values)
display(clean_dataset)
```

## Handling Missing Value -> Impute Data Point Missing Values

```python
def impute_missing_value_with_mean(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].mean(), inplace=True)
    return dataset
    
def impute_missing_value_with_median(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].median(), inplace=True)
    return dataset

dataset_clean_with_mean = impute_missing_value_with_mean(dataset)
dataset_clean_with_median = impute_missing_value_with_mean(dataset)
dataset_clean_with_mean = dataframe_missing_values(dataset_clean_with_mean)
dataset_clean_with_median= dataframe_missing_values(dataset_clean_with_median)

display(dataset_clean_with_mean)
display(dataset_clean_with_median)
```

## Data Pre-Processing : Scalling & Konversi String to Numerik

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

def impute_missing_value_with_mean(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].mean(), inplace=True)
    return dataset

dataset = impute_missing_value_with_mean(dataset)

scalling_columns = dataset.columns[dataset.dtypes == "float64"]
object_columns = dataset.columns[dataset.dtypes == "object"]

scaller = MinMaxScaler()
dataset[scalling_columns] = scaller.fit_transform(dataset[scalling_columns])

from sklearn.preprocessing import LabelEncoder

for column in object_columns:
    LE = LabelEncoder()
    dataset[column] = LE.fit_transform(dataset[column])
    display(LE.classes_)
    display(sorted(dataset[column].unique()))

display(dataset.describe().T[["min","max"]])
```

## Features & Label

```python
X = dataset.drop(["Revenue"], axis=1)
y = dataset["Revenue"]
print("Shape of X = %d Baris & %d Kolom"%X.shape)
print("Shape of y = %d Baris"%y.shape)
```

## Training dan Test Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("Shape of X_train = %d Baris %d Kolom"%X_train.shape)
print("Shape of X_test = %d Baris %d Kolom"%X_test.shape)
print("Shape of y_train = %d Baris"%y_train.shape)
print("Shape of y_test = %d Baris"%y_test.shape)
```

## Training Model: Fit

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model = model.fit(X_train, y_train)
```

## Training Model: Predict

```python
y_pred = model.predict(X_test)
display(y_pred.shape)
```

## Evaluasi Model Performance

```python
from sklearn.metrics import confusion_matrix, classification_report

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("True Negatif = {}\nFalse Positif = {}\nFalse Negatif = {}\nTrue Positif = {}".format(tn, fp, fn, tp))

df_cm = pd.DataFrame(confusion_matrix(y_test, y_pred))

fig, axs = plt.subplots(1, 1, figsize=(12, 5))
color = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(df_cm, annot=True, cmap=color,  fmt="g")
axs.set_title("Confusion Matrix", fontsize=25, fontweight="bold")
axs.set_xlabel("Predicted Value", fontsize=15)
axs.xaxis.set_label_position("top")
axs.set_xticklabels(["Tidak Membeli","Membeli"])
axs.set_yticklabels(["Tidak Membeli","Membeli"])
axs.set_ylabel("Actual Value", fontsize=15)
plt.tight_layout()
plt.show()

cr = classification_report(y_test, y_pred)
print(cr)
print("Accuracy = Jika FN == FP\nF1-Score = Jika FN != FP\nRecall = Jika FP lebih baik daripada FN\nPrecision = Jika FN lebih baik daripada FP")
```
