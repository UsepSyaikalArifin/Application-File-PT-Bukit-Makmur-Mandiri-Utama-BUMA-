## Handling Missing Value -> Drop All Rows Missing Values

```python
def dataframe_missing_values(dataset):
    df_columns = dataset.isnull().sum().index
    df_values = dataset.isnull().sum().values
    df_missing_value = pd.DataFrame({
                                        "Columns" : df_columns,
                                        "Missing_Values" : df_values
                                    })
    return df_missing_value

def dropall_missing_value(dataset):
    dataset_clean = dataset.dropna()
    df_clean_missing_value = dataframe_missing_values(dataset_clean)
    return df_clean_missing_value

missing_values = dataframe_missing_values(dataset)
clean_dataset = dropall_missing_value(dataset)

display(missing_values)
display(clean_dataset)
```

## Handling Missing Value -> Impute Data Point Missing Values

```python
def impute_missing_value_with_mean(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].mean(), inplace=True)
    return dataset
    
def impute_missing_value_with_median(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].median(), inplace=True)
    return dataset

dataset_clean_with_mean = impute_missing_value_with_mean(dataset)
dataset_clean_with_median = impute_missing_value_with_mean(dataset)
dataset_clean_with_mean = dataframe_missing_values(dataset_clean_with_mean)
dataset_clean_with_median= dataframe_missing_values(dataset_clean_with_median)

display(dataset_clean_with_mean)
display(dataset_clean_with_median)
```

## Data Pre-Processing : Scalling & Konversi String to Numerik

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

def impute_missing_value_with_mean(dataset):
    columns_missing_values = list(dataset.columns[dataset.isnull().any()])
    for column in columns_missing_values:
        dataset[column].fillna(dataset[column].mean(), inplace=True)
    return dataset

dataset = impute_missing_value_with_mean(dataset)

scalling_columns = dataset.columns[dataset.dtypes == "float64"]
object_columns = dataset.columns[dataset.dtypes == "object"]

scaller = MinMaxScaler()
dataset[scalling_columns] = scaller.fit_transform(dataset[scalling_columns])

from sklearn.preprocessing import LabelEncoder

for column in object_columns:
    LE = LabelEncoder()
    dataset[column] = LE.fit_transform(dataset[column])
    display(LE.classes_)
    display(sorted(dataset[column].unique()))

display(dataset.describe().T[["min","max"]])
```

## Features & Label

```python
X = dataset.drop(["Revenue"], axis=1)
y = dataset["Revenue"]
print("Shape of X = %d Baris & %d Kolom"%X.shape)
print("Shape of y = %d Baris"%y.shape)
```

## Training dan Test Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("Shape of X_train = %d Baris %d Kolom"%X_train.shape)
print("Shape of X_test = %d Baris %d Kolom"%X_test.shape)
print("Shape of y_train = %d Baris"%y_train.shape)
print("Shape of y_test = %d Baris"%y_test.shape)
```

## Training Model: Fit -> DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model = model.fit(X_train, y_train)
```

## Training Model: Predict

```python
y_pred = model.predict(X_test)
display(y_pred.shape)
```

## Evaluasi Model Performance

```python
from sklearn.metrics import confusion_matrix, classification_report

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("True Negatif = {}\nFalse Positif = {}\nFalse Negatif = {}\nTrue Positif = {}".format(tn, fp, fn, tp))

print("Training Accuracy = {}\nTesting Accuracy ={}".format(model.score(X_train, y_train), model.score(X_test, y_test)))

df_cm = pd.DataFrame(confusion_matrix(y_test, y_pred))

fig, axs = plt.subplots(1, 1, figsize=(12, 5))
color = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(df_cm, annot=True, cmap=color,  fmt="g")
axs.set_title("Confusion Matrix", fontsize=25, fontweight="bold")
axs.set_xlabel("Predicted Value", fontsize=15)
axs.xaxis.set_label_position("top")
axs.set_xticklabels(["Tidak Membeli","Membeli"])
axs.set_yticklabels(["Tidak Membeli","Membeli"])
axs.set_ylabel("Actual Value", fontsize=15)
plt.tight_layout()
plt.show()

cr = classification_report(y_test, y_pred)
print(cr)
print("Accuracy = Jika FN == FP\nF1-Score = Jika FN != FP\nRecall = Jika FP lebih baik daripada FN\nPrecision = Jika FN lebih baik daripada FP")
```

## Modelling -> LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

model_logreg = LogisticRegression(max_iter=300)
model_logreg = model_logreg.fit(X_train, y_train)
y_predict = model_logreg.predict(X_test)

print("Training Accuracy = {}".format(model_logreg.score(X_train, y_train)))
print("Testing Accuracy = {}".format(model_logreg.score(X_test, y_test)))

cm_logreg = confusion_matrix(y_test, y_predict)
df_cm_logreg = pd.DataFrame(cm_logreg)
fig, axs = plt.subplots(1, 1, figsize=(12, 5))
color = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(df_cm_logreg, annot=True, fmt="g", ax=axs, cmap=color)
axs.set_title("Confusion Matrix Model Logistic Regression", fontsize=25, fontweight="bold")
axs.set_xlabel("Predicted Value", fontsize=20)
axs.xaxis.set_label_position("top")
axs.set_ylabel("Actual Value", fontsize=20)
axs.set_xticklabels(["Tidak Membeli","Membeli"])
axs.set_yticklabels(["Tidak Membeli","Membeli"])
plt.tight_layout()
plt.show()

cr_logistic_regression = classification_report(y_test, y_predict)
print("Classification Report Model Logistic Regression : \n{}".format(cr_logistic_regression))
```

## Modelling -> DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report

model_DTC = DecisionTreeClassifier()
model_DTC = model_DTC.fit(X_train, y_train)
y_predict = model_DTC.predict(X_test)

print("Training Accuracy = {}\nTesting Accuracy = {}".format(model_DTC.score(X_train, y_train), model_DTC.score(X_test, y_test)))

cm_DTC = confusion_matrix(y_test, y_predict)
df_cm_DTC = pd.DataFrame(cm_DTC)

fig, axs = plt.subplots(1, 1, figsize=(12, 5))
color = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(df_cm_DTC, annot=True, cmap=color, fmt="g", ax=axs)
axs.set_title("Confusion Matrix of DTC Model", fontsize=25, fontweight="bold")
axs.set_xlabel("Perdicted Value", fontsize=20)
axs.xaxis.set_label_position("top")
axs.set_xticklabels(["Tidak Membeli", "Membeli"])
axs.set_ylabel("Actual Value", fontsize=20)
axs.set_yticklabels(["Tidak Membeli", "Membeli"])
plt.tight_layout()
plt.show()
```

# Unsupervised Learning

```python
import pandas as pd
from IPython.display import display

dataset = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/pythonTutorial/mall_customers.csv")
# dataset.to_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Application File\Application-File-PT-Bukit-Makmur-Mandiri-Utama-BUMA-\Kelas Fundamental\Machine Learning With Python for Beginner\File Dokumentasi\File CSV\mall_customers.csv", index=None, header=True)
# dataset.to_excel(r"D:\Usep Syaikal Arifin\Computer Programming\Application File\Application-File-PT-Bukit-Makmur-Mandiri-Utama-BUMA-\Kelas Fundamental\Machine Learning With Python for Beginner\File Dokumentasi\File Excel\mall_customers.xlsx", index=None, header=True)
print("Shape of Dataset = %d Baris & %d Kolom"%dataset.shape)
print("Informasi Dataset :")
display(dataset.info())
print("Statistik Dasar Dataset :")
display(dataset.describe())
display(dataset.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

features_columns = ["annual_income","spending_score"]
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
for i, column in enumerate(features_columns):
    color = sns.color_palette("Set{}".format(i+2))
    sns.boxplot(x=column, data = dataset, ax=axs[i][0], palette=color)
    sns.histplot(data = dataset, x=column, ax=axs[i][1], palette=color, kde=True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

X = dataset[["annual_income","spending_score"]]

inertia = []
index = []
for k in range(2, 10):
    cluster_model = KMeans(n_clusters=k, random_state=0)
    cluster_model.fit_predict(X)
    inertia_value = cluster_model.inertia_
    inertia.append(inertia_value)
    index.append(k)

cluster_model = KMeans(n_clusters=5, random_state=0)
cluster_model.fit_predict(X)
centroids = cluster_model.cluster_centers_
display(centroids)

color=sns.color_palette("magma")
fig, axs = plt.subplots(2, 1, figsize=(12, 10))
sns.pointplot(x=index, y=inertia, ax=axs[0], markers="D", linestyles="--")
axs[0].set_title("Find Optimal n_clusters", fontsize=25, fontweight="bold")
axs[0].set_xlabel("k CLuster", fontsize=20)
axs[0].set_ylabel("Inertia Value", fontsize=20)
sns.scatterplot(x="annual_income", y="spending_score", data=dataset, ax=axs[1])
sns.scatterplot(x=centroids[:,0], y=centroids[:,1], ax=axs[1], marker="D", s=100)
axs[1].set_title("K Means Clustering", fontsize=25, fontweight="bold")
axs[1].set_xlabel("Annual Income", fontsize=20)
axs[1].set_ylabel("Spending Score", fontsize=20)
plt.tight_layout()
plt.show()
```

```python
labels = cluster_model.fit_predict(X)
df_labels = pd.DataFrame({"Labels":labels})
new_df = dataset.merge(df_labels, left_index=True, right_index=True, how="left")
display(new_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axs = plt.subplots(1, 1, figsize=(12, 6))
color = sns.color_palette("magma")
sns.scatterplot(data=new_df, x="annual_income", y="spending_score", hue="Labels", style="Labels", palette="deep", ax=axs)
sns.scatterplot(x=centroids[:,0], y=centroids[:,1], ax=axs, marker="D", s=200, palette=color)
axs.set_title("K Means Clustering", fontsize=25, fontweight="bold")
axs.set_xlabel("Annual Income", fontsize=20)
axs.set_ylabel("Spending Score", fontsize=20)
plt.tight_layout()
plt.show()
```
